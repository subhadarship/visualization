```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plot
plt.figure(figsize=(5, 4), dpi=160)
plt.plot(np.array(range(5)) - 1, 'go-', label='plain')
plt.plot(np.array(range(5)) - 2, 'ro--', label='dashed')
plt.plot(np.array(range(5)) - 3, 'bo-.', label='dotted dashed')
plt.legend()
plt.show()
```
