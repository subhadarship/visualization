```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create data
x = np.array(range(1, 50 + 1))
ys = np.random.rand(50, 2000)

# compute means
y_means = np.mean(ys, axis=1)

# compute standard deviations
y_stds = np.std(ys, axis=1)
```

Style A

```python
# plot
plt.figure(figsize=(6, 4), dpi=160)
plt.errorbar(x, y_means, y_stds, color='xkcd:red')
plt.show()
```

Style B

```python
# plot
plt.figure(figsize=(6, 4), dpi=160)
plt.plot(x, y_means, color='xkcd:red')
plt.fill_between(x, y_means - y_stds, y_means + y_stds, facecolor='xkcd:red', 
                 edgecolor='None', alpha=0.3)
plt.show()
```
